## Image Recognition with AWS and Python

```python
import cv2
import csv
import boto3
from matplotlib import pyplot as plt

AWS_DEFAULT_REGION='us-west-2'
```

We also add aws defalut region on boto3.client paramter, because if we don't add this it will be error

```python
with open('student02.csv', 'r') as input :
    next(input)
    reader = csv.reader(input)
    for line in reader:
        access_key_id  = line[2]
        secret_access_key = line[3]
```

### P.10.1 Detection Labels

```python
photo = 'kitten.jpg'
img = cv2.imread(photo)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height_shape = img.shape[0]
width_shape = img.shape[1]

client = boto3.client('rekognition',
                      region_name = AWS_DEFAULT_REGION,
                      aws_access_key_id = access_key_id,
                      aws_secret_access_key = secret_access_key)

with open(photo, 'rb') as source_image:
    source_bytes = source_image.read()
    
response = client.detect_labels(Image={'Bytes': source_bytes})

# print(response)
```

```python
print('Detected labels for ' + photo)
print()
for label in response['Labels']:
    print("Label : " + label['Name'])
    print("Confidence : " + str(label['Confidence']))
    print("Instances : ")
    for instance in label["Instances"]:
        print(" Bounding Box")
        print(" Top " + str(instance['BoundingBox']['Top']))
        print(" Left " + str(instance['BoundingBox']['Left']))
        print(" Width " + str(instance['BoundingBox']['Width']))
        print(" Height " + str(instance['BoundingBox']['Height']))
        print(" Confidence " + str(instance['Confidence']))
        
        top = int(instance['BoundingBox']['Top'] * height_shape)
        left = int(instance['BoundingBox']['Left'] * width_shape)
        width = int(instance['BoundingBox']['Width'] * width_shape)
        height = int(instance['BoundingBox']['Height'] * height_shape)
        
        print(" Top ", top)
        print(" Left ", left)
        print(" Width ", width)
        print(" Height ", height)
        
        print()
        img = cv2.rectangle(img, (left,top), (left+width, top+height), (0,255,0), 2)
    
    print("Parents :")
    for parent in label['Parents']:
        print("     " + parent['Name'])
    print("--------------")
    print()
```

```python
plt.imshow(img)
plt.show()
```

### P.10.2 Moderation Label

```python
photo = 'RPG.jpg'
img = cv2.imread(photo)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()

with open(photo, 'rb') as source_image:
    source_bytes = source_image.read()
    
response = client.detect_moderation_labels(Image={'Bytes': source_bytes})

# print(response)
for label in response['ModerationLabels']:
    print("Jenis : " + str(label['Name']))
```

### P.11 Detection Faces

```python
photo = 'bill_and_steve.jpg'
img = cv2.imread(photo)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height_shape = img.shape[0]
width_shape = img.shape[1]

with open(photo, 'rb') as source_image:
    source_bytes = source_image.read()
    
response = client.detect_faces(Image={'Bytes': source_bytes}, Attributes = ['ALL'])

# print(response)

i=1
for person in response['FaceDetails']:
    print(i)
    print(" Bounding Box")
    print(" Top " + str(person['BoundingBox']['Top']))
    print(" Left " + str(person['BoundingBox']['Left']))
    print(" Width " + str(person['BoundingBox']['Width']))
    print(" Height " + str(person['BoundingBox']['Height']))
    print(" Age Low " + str(person['AgeRange']['Low']))
    print(" Age High " + str(person['AgeRange']['High']))
    
    top = int(person['BoundingBox']['Top'] * height_shape)
    left = int(person['BoundingBox']['Left'] * width_shape)
    width = int(person['BoundingBox']['Width'] * width_shape)
    height = int(person['BoundingBox']['Height'] * height_shape)

    print(" Top ", top)
    print(" Left ", left)
    print(" Width ", width)
    print(" Height ", height)

    print()
    img = cv2.rectangle(img, (left,top), (left+width, top+height), (0,255,0), 2)
    i+=1
    
plt.imshow(img)
plt.show()
```
